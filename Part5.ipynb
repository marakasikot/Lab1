```python
import pandas as pd
import matplotlib.pyplot as plt

def parse_csv_for_visualization(file_path='weather.csv'):

    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(file_path)

        # Strip any leading or trailing spaces from column names
        df.columns = df.columns.str.strip()

        # Ensure the date column 'CET' is in datetime format for plotting
        df['CET'] = pd.to_datetime(df['CET'])

        return df

    except FileNotFoundError:
        print("weather.csv file not found.")
        return None

def visualize_data(file_path='weather.csv'):

    # Parse the data
    df = parse_csv_for_visualization(file_path)

    if df is None:
        return

    # Create a figure and axis for plotting
    plt.figure(figsize=(10, 6))

    # Plot max, mean, and min temperatures
    plt.subplot(2, 2, 1)
    plt.plot(df['CET'], df['Max TemperatureC'], label='Max Temperature', color='red', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Mean TemperatureC'], label='Mean Temperature', color='blue', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Min TemperatureC'], label='Min Temperature', color='green', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Temperature (C)')
    plt.title('Temperature over Time')
    plt.legend()

    # Plot max, mean, and min humidity
    plt.subplot(2, 2, 2)
    plt.plot(df['CET'], df['Max Humidity'], label='Max Humidity', color='orange', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Mean Humidity'], label='Mean Humidity', color='purple', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Min Humidity'], label='Min Humidity', color='brown', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Humidity (%)')
    plt.title('Humidity over Time')
    plt.legend()

    # Plot max wind speed
    plt.subplot(2, 2, 3)
    plt.plot(df['CET'], df['Max Wind SpeedKm/h'], label='Max Wind Speed', color='cyan', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed (Km/h)')
    plt.title('Wind Speed over Time')
    plt.legend()

    # Plot total precipitation
    plt.subplot(2, 2, 4)
    plt.bar(df['CET'], df['Precipitationmm'], label='Precipitation', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Precipitation (mm)')
    plt.title('Precipitation over Time')

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plots
    plt.show()

    # Save the plots as images
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 2, 1)
    plt.plot(df['CET'], df['Max TemperatureC'], label='Max Temperature', color='red', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Mean TemperatureC'], label='Mean Temperature', color='blue', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Min TemperatureC'], label='Min Temperature', color='green', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Temperature (C)')
    plt.title('Temperature over Time')
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(df['CET'], df['Max Humidity'], label='Max Humidity', color='orange', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Mean Humidity'], label='Mean Humidity', color='purple', linestyle='-', marker='o')
    plt.plot(df['CET'], df['Min Humidity'], label='Min Humidity', color='brown', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Humidity (%)')
    plt.title('Humidity over Time')
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(df['CET'], df['Max Wind SpeedKm/h'], label='Max Wind Speed', color='cyan', linestyle='-', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed (Km/h)')
    plt.title('Wind Speed over Time')
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.bar(df['CET'], df['Precipitationmm'], label='Precipitation', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Precipitation (mm)')
    plt.title('Precipitation over Time')

    plt.tight_layout()

    # Save the plot as an image (PNG format)
    plt.savefig('weather_visualization.png', format='png')
    plt.savefig('weather_visualization.webp', format='webp')

# Call the visualization function
visualize_data('weather.csv')
```
